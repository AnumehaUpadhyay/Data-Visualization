```python
### Importing Libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#### loading Dataset
iris = pd.read_csv("iris.csv")
```

```python
iris.head(7)
```

```python
### count values in dataset
iris["Species"].value_counts()
```

```python
#### joint plot 
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=7)
```

```python
### FacetGrid
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
### violinplot
sns.violinplot(x="Species", y="SepalLengthCm", data=iris, size=7)
```

```python
sns.violinplot(x="Species", y="SepalWidthCm", data=iris, size=7)
```

```python
###### factorplot
e=sns.factorplot(x="SepalWidthCm", y="SepalLengthCm",col="Species", data=iris, kind="swarm")
```

```python
### boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
sns.violinplot(x="Species", y="PetalWidthCm", data=iris, size=7)
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=7)
```

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
### pairplot
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```
